```python
# Import libraries
from pandas import *
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sio
import math as mth
from sklearn.cluster import KMeans
```

```python
# Load data
import h5py 
dataset = sio.loadmat('ex7data2.mat')
```

```python
# Put X and y data into arrays
Xdata = dataset['X']
```

```python
# Set number of centroids
K = 3
# Set number of initial centroids to test
init_test = 12
```

```python
# Used k-means ++ to optimize convergence, rather than random selection of centroids
kmeans = KMeans(n_clusters=K, init='k-means++', n_init = init_test, max_iter = 20).fit(Xdata)
```

```python
centroids_opt = kmeans.cluster_centers_
```

```python
# Plot final centroids for different numbers of initial centroid tests, using random init
X_plot = plt.scatter(Xdata[:,0], Xdata[:,1])
init_arr = [1, 2, 5, 10, 20]
for i in range(0,5):
    kmeans = KMeans(n_clusters=K, init='random', n_init = init_arr[i], max_iter = 10).fit(Xdata)
    centroids = kmeans.cluster_centers_
    # Plot X with centroids 

    plt.scatter(centroids[:,0], centroids[:,1], label = 'Num Init Centrs = %i' % init_arr[i])
    
plt.legend()

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Centroids for K-means clustering')
```
